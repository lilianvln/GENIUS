# Fonction

```python
# Fonction pour générer une playlist selon les critères donnés
def generate_playlist(df, num_songs, genre=None, artists=None, feats=None, period=None, language=None):
    playlist = df
    
    # Filtrer par genre si fourni
    if genre:
        playlist = playlist[playlist['tag'].isin(genre)]
    
    # Filtrer par artistes si fourni
    if artists:
        playlist = playlist[playlist['artist'].isin(artists)]
    
    # Filtrer par collaborations (feats) si fourni
    if feats:
        playlist = playlist[playlist['features'].apply(lambda x: any(feat in x for feat in feats))]
    
    # Filtrer par période (année) si fournie
    if period:
        start_year, end_year = period
        playlist = playlist[(playlist['year'] >= start_year) & (playlist['year'] <= end_year)]
    
    # Filtrer par langue si fournie
    if language:
        playlist = playlist[playlist['language'].isin(language)]
        
    
    # Limiter au nombre de chansons demandé
    playlist = playlist.head(num_songs)
    
    return playlist
```

# Test

```python
# Exemple d'utilisation
num_songs = 100
genre = ['pop']
artists = []
feats = []
period = (1998, 2024)
language = ['fr']

playlist = generate_playlist(df, num_songs, genre=genre, artists=artists, feats=feats, period=period, language=language)
```
