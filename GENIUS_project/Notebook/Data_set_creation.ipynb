# Cleaning data

## Function

```python
def clean_text(text):
    # Vérifier si la valeur n'est pas vide
    if isinstance(text, str):
        # Mettre en minuscule
        text = text.lower()
        # Supprimer les caractères spéciaux
        text = re.sub(r'[^a-z0-9\s]', '', text)
        # Supprimer les mots vides
        stop_words = set(stopwords.words('english'))  # Remplacer par 'french', etc., si nécessaire
        text = ' '.join(word for word in text.split() if word not in stop_words)
    return text
```

## Cleaning text

```python
df_sample['track_name_cleaned'] = df_sample['track_name'].apply(clean_text)
df_sample['track_artist_cleaned'] = df_sample['track_artist'].apply(clean_text)
df_sample['track_album_name_cleaned'] = df_sample['track_album_name'].apply(clean_text)
df_sample['playlist_name_cleaned'] = df_sample['playlist_name'].apply(clean_text)
df_sample['playlist_genre_cleaned'] = df_sample['playlist_genre'].apply(clean_text)
```

## Clear Year

```python
def extract_year_from_dates(df, column_name):
    # Convertir en datetime tout en gérant les erreurs
    df[column_name] = pd.to_datetime(
        df[column_name], errors='coerce', format='%Y-%m-%d'
    )
    # Extraire l'année et convertir en entier, en remplaçant les NaN par une valeur par défaut (ex: 0 ou None)
    df[column_name] = df[column_name].dt.year.fillna(0).astype(int)
    return df
df_sample = extract_year_from_dates(df_sample, 'track_album_release_date')
```

## Delete NULL

```python
df_sample = df_sample.dropna(subset=['track_name'])
df_sample = df_sample.dropna(subset=['track_artist'])
df_sample = df_sample.dropna(subset=['playlist_name'])
df_sample = df_sample.dropna(subset=['playlist_genre'])
```

## Delete duplicates

```python
df_sample = df_sample.drop_duplicates(subset=['track_artist', 'track_name'], keep='last')
```

# Reset index

# Functions

```python
def get_track_name_by_id(track_id, df_sample):
    """
    Retourne le titre d'une musique à partir de son track_id.

    Parameters:
        track_id (str): L'identifiant de la musique.
        df_sample (pd.DataFrame): Le DataFrame contenant les colonnes 'track_id' et 'track_name'.

    Returns:
        str: Le titre de la musique, ou un message si l'id n'est pas trouvé.
    """
    # Vérifier si l'ID existe dans le DataFrame
    track = df_sample[df_sample['track_id'] == track_id]

    if track.empty:
        return f"Aucun titre trouvé pour l'ID : {track_id}"

    # Retourner le titre
    return track.iloc[0]['track_name']
```

```python
def get_track_id_by_name(track_name, track_artist, df_sample):
    """
    Retourne l'identifiant (track_id) d'une musique à partir de son titre (track_name) et de son artiste (track_artist).
    Si track_name est None, retourne toutes les musiques de l'artiste.

    Parameters:
        track_name (str or None): Le titre de la musique, ou None pour récupérer toutes les musiques de l'artiste.
        track_artist (str): Le nom de l'artiste.
        df_sample (pd.DataFrame): Le DataFrame contenant les colonnes 'track_id', 'track_name', et 'track_artist'.

    Returns:
        pd.DataFrame or str: Un DataFrame des musiques de l'artiste ou un message si aucune correspondance n'est trouvée.
    """
    # Nettoyer le texte des entrées
    track_artist = clean_text(track_artist)

    if track_name is not None:
        track_name = clean_text(track_name)
        # Rechercher une correspondance spécifique dans le DataFrame
        track = df_sample[
            (df_sample['track_name_cleaned'] == track_name) &
            (df_sample['track_artist_cleaned'] == track_artist)
        ]
    else:
        # Rechercher toutes les musiques de l'artiste
        track = df_sample[
            df_sample['track_artist_cleaned'] == track_artist
        ]

    if track.empty:
        return ''

    # Retourner le DataFrame ou l'ID
    return track['track_id'].tolist()
```

```python
def weighted_similarity_score(similarity, popularity, weight_popularity=0.3):
    """
    Combine similarity et popularity dans un score pondéré.
    """
    return similarity * (1 - weight_popularity) + (popularity / 100) * weight_popularity
```

# Model

## Functions

```python
def recommend_songs(track_ids, similarity_matrix, df, top_n=10, weight_popularity=0.3):
    """
    Recommande des chansons similaires à une liste de chansons, avec un score pondéré basé sur la popularité.

    Args:
        track_ids (list): Liste des track_id des musiques pour lesquelles obtenir des recommandations.
        similarity_matrix (ndarray): Matrice de similarité cosinus.
        df (DataFrame): DataFrame contenant les informations sur les musiques.
        top_n (int): Nombre de recommandations à retourner.
        weight_popularity (float): Poids donné à la popularité dans le calcul du score pondéré (entre 0 et 1).

    Returns:
        DataFrame: DataFrame contenant les recommandations triées par score pondéré.
    """

    # Vérifier si tous les track_id existent dans le DataFrame
    if not track_ids:
        return f"Les titres / artistes n'existent pas dans le DataFrame."

    # Récupérer les indices des musiques
    track_indices = [df[df['track_id'] == track_id].index[0] for track_id in track_ids]

    # Calculer la moyenne des similarités pour les musiques sélectionnées
    combined_similarity = np.mean(similarity_matrix[track_indices], axis=0)

    # Trouver les indices des musiques les plus similaires (hors des musiques d'entrée)
    similar_indices = np.argsort(combined_similarity)[::-1]
    similar_indices = [idx for idx in similar_indices if idx not in track_indices][:top_n]

    # Récupérer les scores de similarité
    similar_scores = combined_similarity[similar_indices]

    # Créer un DataFrame pour les résultats
    recommendations = df.iloc[similar_indices][['track_name', 'track_artist', 'track_popularity']].copy()
    recommendations['similarity_score'] = similar_scores

    # Calculer un score pondéré basé sur la popularité
    recommendations['weighted_score'] = recommendations.apply(
        lambda row: weighted_similarity_score(row['similarity_score'], row['track_popularity'], weight_popularity),
        axis=1
    )

    # Trier par score pondéré
    recommendations = recommendations.sort_values(by='weighted_score', ascending=False)

    return recommendations
```

```python
# Calculer les similarités cosinus entre les chansons
feature_vectors = df_sample[features].values
similarity_matrix = cosine_similarity(feature_vectors)

# Exemple d'utilisation
track_ids = [
    get_track_id_by_name(None, 'tyler the creator', df_sample),
]
track_ids = list(itertools.chain.from_iterable(track_ids))
recommendations = recommend_songs(track_ids, similarity_matrix, df_sample, top_n=10)
```
