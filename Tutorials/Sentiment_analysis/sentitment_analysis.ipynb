```python
from transformers import pipeline

# Présentation
def presentation():
    print("Bonjour, je m'appelle Lilian Valin, et je travaille sur un projet nommé GENIUS.")
    print("Je vais vous expliquer ce qu'est l'analyse des sentiments (sentiment analysis) en NLP.")
    print("\nL'analyse des sentiments consiste à déterminer automatiquement l'opinion ou l'émotion exprimée dans un texte.")
    print("Elle est utilisée pour comprendre si un texte est positif, négatif ou neutre, comme dans les avis clients ou les posts sur les réseaux sociaux.\n")

# Exemple d'analyse des sentiments
def sentiment_analysis_example(text):
    # Charger un pipeline d'analyse des sentiments avec un modèle spécifique
    sentiment_analyzer = pipeline(
        "sentiment-analysis",
        model="distilbert-base-uncased-finetuned-sst-2-english",  # Spécification explicite
        device=0  # Utilise le GPU si disponible (0 pour GPU, -1 pour CPU)
    )
    # Obtenir le résultat
    result = sentiment_analyzer(text)
    return result

# Appel des fonctions
if __name__ == "__main__":
    presentation()
    texte = input("Entrez un texte pour analyser ses sentiments :\n")
    resultat = sentiment_analysis_example(texte)

    # Afficher les résultats
    print("\nRésultat de l'analyse des sentiments :")
    for item in resultat:
        print(f"- Sentiment : {item['label']} (Score : {item['score']:.2f})")
```
