```python
from collections import defaultdict
```

# Functions

```python
def get_pairs(text):
    pairs = defaultdict(int)
    for i in range(len(text) - 1):
        pairs[(text[i], text[i+1])] += 1
    print('Frequency of Pairs : ', pairs)
    return pairs
```

```python
def merge_pairs(text, pairs):
    new_text = []
    i = 0
    while i < len(text):
        if i < len(text) - 1 and (text[i], text[i+1]) == pairs:
            new_text.append(''.join(pairs)+' ')
            i+=2
            print("Merged Pair : ", pairs)
        else:
            new_text.append(text[i])
            i+=1
            print('Merged Pair : ', pairs)
    return new_text
```

```python
def byte_pair_encoding(text, num_merges):
    text = list(text)
    for _ in range(num_merges):
        pairs = get_pairs(text)
        if not pairs:
            break
        most_frequent_pair = max(pairs, key=pairs.get)
        text = merge_pairs(text, most_frequent_pair)
    return text
```

# Example

```python
input_text = ['a','a','a','b','d','a','a','a','b','a','c']
num_merge = 10
result = byte_pair_encoding(input_text, num_merge)
print("Final result", result)
```
